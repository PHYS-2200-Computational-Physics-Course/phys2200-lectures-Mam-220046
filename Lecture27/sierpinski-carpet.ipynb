
## Sierpinski carpet

```julia

using PyPlot
```

```julia

"""
    plot_triangle(A, B, C)

Plot a triangle given the coordinates of its vertices
"""
function plot_triangle(A, B, C)
    x = [A[1], B[1], C[1], A[1]]
    y = [A[2], B[2], C[2], A[2]]
    plot(x, y, color="blue")
    return nothing
end
```

```julia

"""
    plot_sierpinski_triangle(A, B, C, depth)

Plot a Sierpinski triangle given the coordinates of its outer vertices and 
the subdivision depth
"""
function plot_sierpinski_triangle(A, B, C, depth)
    if depth == 0
        plot_triangle(A, B, C)
    else
        plot_sierpinski_triangle(A, (A+B)/2, (A+C)/2, depth-1)
        plot_sierpinski_triangle((A+B)/2, B, (B+C)/2, depth-1)
        plot_sierpinski_triangle((A+C)/2, (B+C)/2, C, depth-1)
    end
    return nothing
end
```

```julia

plot_sierpinski_triangle([0., 0.], [0.5, 0.5*sqrt(3.0)], [1., 0.], 5)
axis(false)
axis("square");
```

```julia

"""
    plot_square(A, B)

Plot a square given the coordinates of its 2 diagonally-opposite vertices
"""
function plot_square(A, B)
    x = [A[1], B[1], B[1], A[1], A[1]]
    y = [A[2], A[2], B[2], B[2], A[2]]
    plot(x, y, color="blue")
    return nothing
end
```

```julia

"""
    plot_sierpinski_square(A, B, depth)

Plot a Sierpinski square given the coordinates of its 2 diagonally-opposite 
outer vertices and the subdivision depth
"""
function plot_sierpinski_square(A, B, depth)
    if depth == 0
        plot_square(A, B)
    else
        dx = [(B[1] - A[1])/3, 0.]
        dy = [0., (B[2] - A[2])/3]
        plot_sierpinski_square(A,        A + dx + dy,   depth-1)
        plot_sierpinski_square(A + dy,   A + dx + 2*dy, depth-1)
        plot_sierpinski_square(A + 2*dy, A + dx + 3*dy, depth-1)
        plot_sierpinski_square(A + dx,        A + 2*dx + dy,   depth-1)
        plot_sierpinski_square(A + dx + 2*dy, A + 2*dx + 3*dy, depth-1)
        plot_sierpinski_square(A + 2*dx,        A + 3*dx + dy,   depth-1)
        plot_sierpinski_square(A + 2*dx + dy,   A + 3*dx + 2*dy, depth-1)
        plot_sierpinski_square(A + 2*dx + 2*dy, A + 3*dx + 3*dy, depth-1)
    end
    return nothing
end
```

```julia

plot_sierpinski_square([0., 0.], [1., 1.], 4)
axis(false)
axis("square");
```
